```python
from keras.models import Model, Input, load_model
from load_data import load_data_keras
import pandas as pd
from matplotlib.image import imread
import numpy as np

from ipynb.fs.defs.DenseNet201_pretrained_cnn import DenseNet201_cnn
from ipynb.fs.defs.conv_pool_cnn import conv_pool_cnn
```

```python
def load_label_names():
    return ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
input_shape = (32, 32, 3)
model_input = Input(shape=input_shape)
```

```python
conv_pool_cnn_model = conv_pool_cnn(model_input)
```

```python
conv_pool_cnn_model.load_weights('saved_models\\Conv_pool_4.h5')
```

```python
test_images = []
for i in range(1,300_001):
    test_images.append(imread('data\\test\\'+str(i)+'.png'))
    if i%1000 == 0:
        print(f"Progress: {i}/300 000")
```

```python
predictions = np.array([])
for i in range(0,300_000):
    X_test = np.array(test_images[i]).reshape(-1, 32, 32, 3)
    prediction = conv_pool_cnn_model.predict(X_test)
    predictions = np.append(predictions, np.argmax(prediction,axis=1))
    if i%1000 == 0:
        print(f"Progress: {i}/300000")
```

```python
labels = load_label_names()
predicted_labels = [labels[x] for x in predictions]
predictions_df = pd.DataFrame({'id': range(1,len(prediction_labels)+1), 'label': predicted_labels})
predictions_df.to_csv("kaggle_conv_pool_1.csv", index = None)
predictions2_df = pd.DataFrame(conv_pool_cnn_model.predict(X_test))
predictions2_df.to_csv("proba_conv_pool_1.csv", index = None)
```
